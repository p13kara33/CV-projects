```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df = pd.read_excel(r'C:\Users\Giorgis\Desktop\Projects\tableau\toolsprop.xlsx')
df
```

```python
tools_names = df['Tool Name'].values.tolist()
project_num = df['Projects'].values.tolist()
```

```python
plt.figure(figsize=(10,10))

# Pie chart, where the slices will be ordered and plotted counter-clockwise:
labels = tools_names
sizes = project_num
explode = (0, 0, 0, 0, 0, 0, 0)  # only "explode" the 2nd slice (i.e. 'Hogs')


plt.pie(sizes, explode=explode, labels=labels, autopct='%1.1f%%',
        shadow=True, startangle=90, textprops={'fontsize': 14, 'family':'Verdana'})
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.


plt.show()
```
