```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df = pd.read_excel(r'C:\Users\Giorgis\Desktop\Projects\tableau\experience.xlsx')
df
```

```python
dfs = {}
for tool in df["Tools"].unique():
    dfs[tool] = df[(df['Tools'] == tool)]
```

```python
df['Tools'].value_counts().tolist()
```

```python
tools_names = df['Tools'].value_counts().index
months = df['Tools'].value_counts().tolist()
```

```python
plt.figure(figsize=(15,10))
sns.set_style('whitegrid')

ax = sns.lineplot(x=tools_names, y=months)


ax.set_xlabel('Tools', size=20, fontweight='bold')
ax.set_ylabel('Months of experience' ,size=20, fontweight='bold')
ax.set_title('Experience in Tools', size=30, fontweight='bold')



plt.setp(ax.get_xticklabels(), rotation=0, size=15)

plt.setp(ax.get_yticklabels(), rotation=0, size=15)

plt.tight_layout()
sns.despine()

# plt.savefig('lineplot_tools_exp.jpeg')
```

```python
plt.figure(figsize=(10,10))

plt.pie(months, labels=tools_names, autopct='%1.1f%%',
        shadow=True, startangle=90, textprops={'fontsize': 14, 'family':'Verdana'})
# plt.axis('equal')?


plt.savefig('piechart_tools_exp.jpeg')
```

```python
project_names = df['Type of Project'].value_counts().index
pjmonths = df['Type of Project'].value_counts().tolist()
```

```python
plt.figure(figsize=(18,10))
sns.set_style('whitegrid')

ax = sns.lineplot(x=project_names, y=pjmonths)


ax.set_xlabel('Type of Project', size=18, fontweight='bold')
ax.set_ylabel('Months of experience' ,size=20, fontweight='bold')
ax.set_title('Experience among various fields', size=30, fontweight='bold')



plt.setp(ax.get_xticklabels(), rotation=0, size=15)

plt.setp(ax.get_yticklabels(), rotation=0, size=15)

plt.tight_layout()
sns.despine()

plt.savefig('lineplot_field_exp.jpeg')
```

```python
plt.figure(figsize=(10,10))

plt.pie(pjmonths, labels=project_names, autopct='%1.1f%%',
        shadow=True, startangle=90, textprops={'fontsize': 14, 'family':'Verdana'})
# plt.axis('equal')?


plt.savefig('piechart_field_exp.jpeg')
```
